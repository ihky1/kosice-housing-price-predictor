```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/processed/kosice_housing_prices_data.csv")

df.info()
df.describe()
```

## Clean up outliers in price and area

```python
area_Q1 = df["area"].quantile(0.25)
area_Q3 = df["area"].quantile(0.75)
area_IQR = area_Q3 - area_Q1

price_Q1 = df["price"].quantile(0.25)
price_Q3 = df["price"].quantile(0.75)
price_IQR = price_Q3 - price_Q1

df = df[
    (df["area"] >= area_Q1 - 1.5 * area_IQR) 
    & (df["area"] <= area_Q3 + 1.5 * area_IQR) 
    & (df["price"] >= price_Q1 - 1.5 * price_IQR) 
    & (df["price"] <= price_Q3 + 1.5 * price_IQR)
]
```

```python
average_price = df['price'].mean()
display(average_price)
```

```python
average_price_by_number_of_rooms = df.groupby("rooms")["price"].mean()
display(average_price_by_number_of_rooms)

average_price_by_number_of_rooms.plot(kind='bar', color='blue')
plt.title("Average rental price by number of rooms")
plt.xlabel("Number of rooms")
plt.ylabel("Price")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
average_price_by_district = df.groupby("district")["price"].mean().sort_values()
display(average_price_by_district)

plt.figure(figsize=(14, 6))
average_price_by_district.plot(kind='bar', color='blue')
plt.title("Average rental price by district")
plt.xlabel("District")
plt.ylabel("Price")
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
average_price_by_county = df.groupby("county")["price"].mean().sort_values()
display(average_price_by_county)

average_price_by_county.plot(kind='bar', color='blue')
plt.title("Average rental price by county")
plt.xlabel("County")
plt.ylabel("Price")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df["area"], df["price"], alpha=0.6)
plt.title("Price vs Area")
plt.xlabel("Area (m²)")
plt.ylabel("Price (€/mes.)")
plt.show()
```
