### Load the dataset and display its structure and summary statistics

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/processed/kosice_housing_prices_data.csv")

df.info()
df.describe()
```

### Remove outliers from price and area columns using the IQR method

```python
area_Q1 = df["area"].quantile(0.25)
area_Q3 = df["area"].quantile(0.75)
area_IQR = area_Q3 - area_Q1

price_Q1 = df["price"].quantile(0.25)
price_Q3 = df["price"].quantile(0.75)
price_IQR = price_Q3 - price_Q1

df = df[
    (df["area"] >= area_Q1 - 1.5 * area_IQR) 
    & (df["area"] <= area_Q3 + 1.5 * area_IQR) 
    & (df["price"] >= price_Q1 - 1.5 * price_IQR) 
    & (df["price"] <= price_Q3 + 1.5 * price_IQR)
]
```

### Calculate and display the overall average rental price

```python
average_price = df['price'].mean()
display(average_price)
```

### Visualize average rental price grouped by number of rooms

```python
average_price_by_number_of_rooms = df.groupby("rooms")["price"].mean()
display(average_price_by_number_of_rooms)

average_price_by_number_of_rooms.plot(kind='bar', color='blue')
plt.title("Average rental price by number of rooms")
plt.xlabel("Number of rooms")
plt.ylabel("Price")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

### Visualize average rental price grouped by district

```python
district_counts = df["district"].value_counts()
valid_districts = district_counts[district_counts > 10].index
df_districts_filtered = df[df["district"].isin(valid_districts)]
average_price_by_district = df_districts_filtered.groupby("district")["price"].mean().sort_values()
display(average_price_by_district)

plt.figure(figsize=(14, 6))
average_price_by_district.plot(kind='bar', color='blue')
plt.title("Average rental price by district")
plt.xlabel("District")
plt.ylabel("Price")
plt.xticks(rotation=45, ha="right")
plt.show()
```

### Visualize average rental price grouped by county

```python
county_counts = df["county"].value_counts()
valid_counties = county_counts[county_counts > 10].index
df_counties_filtered = df[df["county"].isin(valid_counties)]
average_price_by_county = df_counties_filtered.groupby("county")["price"].mean().sort_values()
display(average_price_by_county)

average_price_by_county.plot(kind='bar', color='blue')
plt.title("Average rental price by county")
plt.xlabel("County")
plt.ylabel("Price")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

### Scatter plot showing the relationship between apartment area and rental price

```python
plt.figure(figsize=(10, 6))
plt.scatter(df["area"], df["price"], alpha=0.6)
plt.title("Price vs Area")
plt.xlabel("Area (m²)")
plt.ylabel("Price (€/mes.)")
plt.show()
```
